```python
df['Population'] = df['Population'].str.replace(',', '')
df['Population'] = df['Population'].fillna(0)
df['Population'] = df['Population'].apply(int)
```

```python
df['population_squared'] = (df['Population'] * df['Population'])
df['murder_yes'] = np.where(df.iloc[:, 3]>=1, 1, 0)
df['robbery_yes'] = np.where(df.iloc[:, 3]>=1, 1, 0)
```
